```python
#pakistan Scorecard
col_1={"Batsman_name":[],
       "Runs":[],
       "Balls":[],
       "4s":[],
       "6s":[],
       "S/R":[]}

for i in data["data"]["scorecard"][0]["batting"]:
    batsman_name=i['batsman']["name"]
    runs=i['r']
    balls=i["b"]
    fours=i["4s"]
    six=i["6s"]
    s_r=i["sr"]
    col_1["Batsman_name"].append(batsman_name)
    col_1["Runs"].append(runs)
    col_1["Balls"].append(balls)
    col_1["4s"].append(fours)
    col_1["6s"].append(six)
    col_1["S/R"].append(s_r)
    print(batsman_name,"run",runs,balls,fours,six,s_r)
    
```

```python
df_pk=pd.DataFrame(col_1)
df_pk
```
